# Load & Clean Data

```python
df = load_data()

print("Raw data shape:", df.shape)
print("\nFirst 3 rows:")
df.head(3)
```

# Encode Data

```python
df = clean_data(df)
df, encoding_maps = encode_data(df)  # unpack both return values

print("Cleaned & encoded shape:", df.shape)
```

```python
# See the mapping for Suggested Job Role
print("Suggested Job Role mapping:")
for number, label in encoding_maps['Suggested Job Role'].items():
    print(f"  {number} → {label}")
```

```python
for col, mapping in encoding_maps.items():
    print(f"\n--- {col} ---")
    for number, label in mapping.items():
        print(f"  {number} → {label}")
```

```python
# Helper function — pass any column name and number to get the label back
def decode(column_name, number):
    return encoding_maps[column_name][number]

# Example usage
print(decode('Suggested Job Role', 0))   # → Software Developer
print(decode('Suggested Job Role', 1))   # → Data Scientist
```

# Split Features & Target

```python
X, y = split_features_target(df)

print("Features shape:", X.shape)
print("Target shape  :", y.shape)
print("\nJob role distribution:")
print(y.value_counts())
```

# Train/Test Split

```python
X_train, X_test, y_train, y_test = split_train_test(X, y)

print(f"Training samples : {len(X_train)}")
print(f"Test samples     : {len(X_test)}")
```

# Train Model

```python
model = train_model(X_train, y_train)
print("Model trained successfully!")
```

# Evaluate Model

```python
evaluate_model(model, X_test, y_test)
```

#  Top-3 Predictions

```python
labels, probs = get_top_k_predictions(model, X_test.iloc[:5], k=3)

print("Top-3 predicted job roles for first 5 test samples:")
print("-" * 52)
for i, (row_labels, row_probs) in enumerate(zip(labels, probs)):
    print(f"\nSample {i+1}:")
    for rank, (label, prob) in enumerate(zip(row_labels, row_probs), 1):
        print(f"  {rank}. {label:<35} {prob:.2%}")
```
