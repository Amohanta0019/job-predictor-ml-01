# Imports

```python
from quiz_preprocessing import load_quiz_data, preprocess_quiz_data, split_quiz_features_target, split_quiz_train_test, SKILL_COLUMNS
from quiz_model import train_quiz_model
from quiz_predict import predict_from_quiz
from career_info import load_career_info, get_career_path
from evaluate import evaluate_model   # reuse the same evaluate from Part 2
```

#  Load & Preprocess

```python
df = load_quiz_data()
df = preprocess_quiz_data(df)

print("Shape:", df.shape)
print("\nJob role distribution:")
print(df['Role'].value_counts())
```

#  Split & Train

```python
X, y = split_quiz_features_target(df)
X_train, X_test, y_train, y_test = split_quiz_train_test(X, y)

print(f"Training samples : {len(X_train)}")
print(f"Test samples     : {len(X_test)}")

model = train_quiz_model(X_train, y_train)
print("\nQuiz model trained!")
```

# Evaluate

```python
evaluate_model(model, X_test, y_test)
```

#  Simulate a Quiz & Predict

```python
# Simulate a user filling out the quiz
user_answers = {
    'Database Fundamentals':          'Poor',
    'Computer Architecture':          'Poor',
    'Distributed Computing Systems':  'Poor',
    'Cyber Security':                 'Poor',
    'Networking':                     'Poor',
    'Software Development':           'Not Interested',
    'Programming Skills':             'Poor',
    'Project Management':             'Poor',
    'Computer Forensics Fundamentals':'Poor',
    'Technical Communication':        'Poor',
    'AI ML':                          'Not Interested',
    'Software Engineering':           'Not Interested',
    'Business Analysis':              'Poor',
    'Communication skills':           'Poor',
    'Data Science':                   'Not Interested',
    'Troubleshooting skills':         'Poor',
    'Graphics Designing':             'Poor'
}

results = predict_from_quiz(model, user_answers, k=3)

print("Top 3 Suggested Roles Based on Your Skills:")
print("-" * 45)
for rank, (role, prob) in enumerate(results, 1):
    print(f"  {rank}. {role:<35} {prob:.2%}")
```
